## 1.2 å¦‚ä½•ç”Ÿæˆ Var

é¦–å…ˆï¼Œæˆ‘ä»¬åœ¨ Python ä¸­å¯¼å…¥è®¡å›¾ï¼š

```python
import jittor as jt
```

**ä» â€œæ™®é€šæ•°æ®â€ å¯¼å…¥ç”Ÿæˆ Var**

```python
data = [[1, 2],[3, 4],[5, 6]]                   # ç”Ÿæˆä¸€ä¸ªæ™®é€šçš„åˆ—è¡¨
print("data: \n", data)

var_data = jt.array(data)                         # é€šè¿‡åˆ—è¡¨ç”Ÿæˆ Var
print("var_data: \n", var_data)

print("var_dataçš„ç±»å‹æ˜¯:", type(var_data))                          # æ‰“å°ç±»å‹ï¼Œå¯ä»¥å‘ç°ç”Ÿæˆçš„æ˜¯jittor.Var
var_data = jt.Var(data)                        # å½“ç„¶ï¼Œjittor.Varä¹Ÿå¯ä»¥ç›´æ¥ä½¿ç”¨
```

**ä» â€œNumPy æ•°ç»„â€ å¯¼å…¥ç”Ÿæˆ Var**

```python
import numpy as np
np_array = np.array([[1, 2],[3, 4],[5, 6]])      # ç”Ÿæˆä¸€ä¸ª NumPy çš„æ•°ç»„
print("np_array: \n", np_array)

var_np_array=jt.array(np_array)                    # é€šè¿‡ NumPy çš„æ•°ç»„ç”Ÿæˆ Var
print("var_np_array: \n", var_np_array)
```

**ä» â€œå…¶ä»– Var æ•°æ®â€ å¯¼å…¥ç”Ÿæˆæ–°çš„ Var**

```python
# å‡†å¤‡ä¸€ä¸ªåŸå§‹çš„ Var
var_data = jt.Var([[1, 2],[3, 4],[5, 6]])

# ç”Ÿæˆä¸€ä¸ªå½¢çŠ¶ç›¸åŒï¼Œå…ƒç´ å…¨ä¸º 0 çš„ Var
var_zeros = jt.zeros_like(var_data) 
print("Zeros Var: \n", var_zeros)

# ç”Ÿæˆä¸€ä¸ªå½¢çŠ¶ç›¸åŒï¼Œå…ƒç´ å…¨ä¸º 1 çš„ Var
var_ones = jt.ones_like(var_data) 
print("Ones Var: \n", var_ones)

# ç”Ÿæˆä¸€ä¸ªå½¢çŠ¶ç›¸åŒï¼Œå¹¶æŒ‡å®šå…ƒç´ å€¼çš„ Var
var_full = jt.full_like(var_data, 5)
print("full Var: \n", var_full)

# ç”Ÿæˆä¸€ä¸ªå½¢çŠ¶ç›¸åŒï¼Œå…ƒç´ ä¸ºéšæœºæ•°çš„ Var
var_rand = jt.rand(var_data.shape, dtype=jt.float)    # é‡å†™å…ƒç´ æ•°æ®ç±»å‹
print("Random Var: \n", var_rand)
```

**ç”Ÿæˆä¸€äº›ç‰¹å®šçš„ Var**

```python
shape = (3,4,)                                 # æŒ‡å®šå°†è¦ç”Ÿæˆ Var çš„å½¢çŠ¶
rand_var = jt.rand(shape)                      # ç”Ÿæˆå…ƒç´ ä¸ºéšæœºæ•°çš„ Var
ones_var = jt.ones(shape)                      # ç”Ÿæˆå…ƒç´ å…¨ä¸º 1 çš„ Var
zeros_var = jt.zeros(shape)                    # ç”Ÿæˆå…ƒç´ å…¨ä¸º 0 çš„ Var

print("Random Var: \n", rand_var)
print("Ones Var: \n", ones_var)
print("Zeros Var: \n", zeros_var)

arange_var1 = jt.arange(9)                     # ç”Ÿæˆå…ƒç´ ä¸ºç­‰å·®æ•°ç»„çš„ä¸€ç»´ Var (åªæŒ‡å®šä¸€ä¸ªç»“æŸå‚æ•°)
arange_var2 = jt.arange(0, 9, 2)               # ç”Ÿæˆå…ƒç´ ä¸ºç­‰å·®æ•°ç»„çš„ä¸€ç»´ Var (æŒ‡å®šèµ·å§‹ã€ç»“æŸã€æ­¥é•¿)
print("Arange Var 1: \n", arange_var1)
print("Arange Var 2: \n", arange_var2)
```

## 1.4 Var çš„å¸¸è§æ“ä½œ

**é‡å¡‘å½¢çŠ¶ Reshape**

```python
var1 = jt.arange(27)                         # åˆ›å»ºä¸€ä¸ªå…ƒç´ ä¸º 0-26 çš„ä¸€ç»´ Var1
print("Var1: \n", var1)
print("Shape of Var1: \n", var1.shape)

var2 = var1.reshape(3,3,3)                   # å°† Var1 é‡å¡‘å½¢çŠ¶ä¸º 3*3*3 çš„ Var2
print("Var2: \n", var2)
print("Shape of Var2: \n", var2.shape)

var3 = var2.reshape(3,-1)                    # å‚æ•°å€¼ä¸º -1ï¼Œä»£è¡¨å°†æ ¹æ®å…¶ä»–å½¢çŠ¶å‚æ•°è‡ªåŠ¨è®¡ç®—å‡ºåˆé€‚çš„å€¼ï¼Œå†è¿›è¡Œé‡å¡‘å½¢çŠ¶
print("Var3: \n", var3)
print("Shape of Var3: \n", var3.shape)
```

**è½¬ç½® Transpose**

* çŸ©é˜µè½¬ç½®

```python
# ï¼ˆæç¤ºï¼šè¿™é‡Œæˆ‘ä»¬ç”¨ arange åˆå§‹åŒ–ä¸€ä¸ª 0 åˆ° 11 çš„ä¸€ç»´ Varï¼Œç„¶åé€šè¿‡ reshape å°†å…¶è½¬åŒ–ä¸º 4*3 çš„ Var1ã€‚ï¼‰
var1 = jt.arange(12).reshape(4,3)                           # åˆ›å»ºä¸€ä¸ª 4*3 çš„äºŒç»´çŸ©é˜µ Var1
print("Var1: \n", var1)
print("Shape of Var1: \n", var1.shape)

var1_t = var1.transpose()                                   # çŸ©é˜µè½¬ç½®æˆ 3*4 çš„äºŒç»´çŸ©é˜µ Var1 Transpose
print("Var1 Transpose: \n", var1_t)
print("Shape of Var1 Transpose: \n", var1_t.shape)
```

* æŒ‡å®šè½´è½¬ç½®ï¼ˆè¾ƒéš¾ï¼‰

```python
var1 = jt.arange(24).reshape(2,3,4)                        # åˆ›å»ºä¸€ä¸ª 2*3*4 çš„ä¸‰ç»´ Var1
print("Var1: \n", var1)                                    # å…¶å«ä¹‰ä¸ºï¼Œç¬¬ä¸€ä¸ªè½´ä¸Š 2 ä¸ªå…ƒç´ ï¼›ç¬¬äºŒä¸ªè½´ä¸Š 3 ä¸ªå…ƒç´ ï¼›ç¬¬ä¸‰ä¸ªè½´ä¸Š 4 ä¸ªå…ƒç´ ã€‚
print("Shape of Var1: \n", var1.shape)

var1_t = var1.transpose(0,2,1)                             # æ ¹æ®è½´ç´¢å¼•è½¬ç½®ã€‚ä¸‰ç»´è½´ç´¢å¼•ä¸º 0ï¼Œ1ï¼Œ2ï¼Œè°ƒæ•´ç´¢å¼•é¡ºåºå³ä»£è¡¨å¯¹åº”è½´è°ƒæ•´é¡ºåºã€‚
print("Var1 Transpose: \n", var1_t)                        # ä»é»˜è®¤çš„â€œ0ï¼Œ1ï¼Œ2â€è°ƒæ•´ä¸ºâ€œ0ï¼Œ2ï¼Œ1â€ï¼Œå³ä»£è¡¨åä¸¤è½´è¿›è¡Œäº’æ¢ã€‚
print("Shape of Var1 Transpose: \n", var1_t.shape)         # äº’æ¢åï¼Œç¬¬ä¸€ä¸ªè½´ä¸Š 2 ä¸ªå…ƒç´ ï¼›ç¬¬äºŒä¸ªè½´ä¸Š 4 ä¸ªå…ƒç´ ï¼›ç¬¬ä¸‰ä¸ªè½´ä¸Š 3 ä¸ªå…ƒç´ ã€‚
```

**åˆ‡ç‰‡å’Œç´¢å¼•**

```python
var = jt.arange(9).reshape(3,3)                     # åˆå§‹åŒ–ä¸€ä¸ª 3*3 çš„ Var
print(1, "Var: \n", var)

var[:,1] = 0                                        # åˆ©ç”¨åˆ‡ç‰‡ï¼Œå°†ä»»æ„è¡Œçš„ç¬¬äºŒä¸ªå…ƒç´ ï¼ˆå³ç¬¬äºŒåˆ—ï¼‰é‡ç½®ä¸º 0
print(2, "Var: \n", var)

line_2 = var[1]                                     # åˆ©ç”¨ç´¢å¼•ï¼Œæ‹¿åˆ°ç¬¬äºŒè¡Œå…ƒç´ 
print(3, "Line 2 in Var: \n", line_2)
```

**åŸåœ°æ“ä½œ**  
åœ¨è®¡å›¾ä¸­ï¼Œæˆ‘ä»¬ä½¿ç”¨ assign() å‡½æ•°å®ç°åœ¨åŸå˜é‡ä¸Šç›´æ¥æ“ä½œã€‚

```python
var0 = jt.arange(6).reshape(2,3)              # åˆå§‹åŒ–ä¸€ä¸ª 2*3 çš„ Var0
print(0, "Var0: \n", var0)                    # æ‰“å°æŸ¥çœ‹ Var0

var1 = var0.add(1)                            # ä½¿ Var1 ç­‰äº Var0 ä¸Šæ¯ä¸ªå…ƒç´ åŠ  1
print(1, "Var1: \n", var1)                    # æ‰“å°å‘ç° Var1 èµ‹å€¼æˆåŠŸ
print(1, "Var0: \n", var0)                    # ä½† Var0 å¹¶æ²¡æœ‰å‘ç”Ÿå˜åŒ–

var2 = var0.assign(var0.add(1))               # åˆ©ç”¨ assign() å‡½æ•°ï¼Œä½¿ Var2 ç­‰äº Var0 ä¸Šæ¯ä¸ªå…ƒç´ åŠ  1
print(2, "Var2: \n", var2)                    # æ‰“å°å‘ç° Var2 èµ‹å€¼æˆåŠŸ
print(2, "Var0: \n", var0)                    # åŒæ—¶ Var0 ä¹Ÿå‘ç”Ÿäº†æ”¹å˜

var3 = var0.transpose()                       # ä½¿ Var3 ç­‰äº Var0 çš„è½¬ç½®
print(3, "Var3: \n", var3)                    # æ‰“å°å‘ç° Var3 èµ‹å€¼æˆåŠŸ
print(3, "Var0: \n", var0)                    # ä½† Var0 å¹¶æ²¡æœ‰å‘ç”Ÿå˜åŒ–

var4 = var0.assign(var0.transpose())          # åˆ©ç”¨ assign() å‡½æ•°ï¼Œä½¿ Var4 ç­‰äº Var0 çš„è½¬ç½®
print(4, "Var4: \n", var4)                    # æ‰“å°å‘ç° Var4 èµ‹å€¼æˆåŠŸ
print(4, "Var0: \n", var0)                    # åŒæ—¶ Var0 ä¹Ÿå‘ç”Ÿäº†æ”¹å˜
```

**Var çš„åŠ æ³•å’Œè¿æ¥**

```python
# åˆå§‹åŒ–ä¸¤ä¸ªVar
var1 = jt.arange(9).reshape(3,3)
var2 = jt.ones((3,3))
print("var1: \n", var1)
print("var2: \n", var2)
```

```python
# Var ä¸å®æ•°ç›¸åŠ 
var3 = var1 + 3
print("var3: \n", var3)

# ä¸¤ä¸ª Var ç›¸åŠ 
var4 = var1 + var2
print("var4: \n", var4)

# è¿æ¥ä¸¤ä¸ª Var 
var5 = jt.concat([var1, var2], dim=1)
print("var5: \n", var5)
```

**Var çš„ä¹˜æ³•ï¼šæ ‡é‡ç›¸ä¹˜ / çŸ©é˜µç›¸ä¹˜**

```python
# åˆå§‹åŒ–ä¸¤ä¸ªVar
var1 = jt.arange(9).reshape(3,3)
var2 = jt.ones((3,3))
print("var1: \n", var1)
print("var2: \n", var2)
```

* **Var ä¸å®æ•°ç›¸ä¹˜ï¼š**

```python
# Var ä¸å®æ•°ç›¸ä¹˜
var3 = var1 * 3
print("var3: \n", var3)
```

* **ä¸¤ä¸ª Var æ ‡é‡ç›¸ä¹˜ï¼š**

```python
# æ ‡é‡ç›¸ä¹˜ æ–¹å¼ä¸€
var4 = var1 * var2
print("var4: \n", var4)

# æ ‡é‡ç›¸ä¹˜ æ–¹å¼äºŒ
var5 = var1.multiply(var2)
print("var5: \n", var5)

# æ ‡é‡ç›¸ä¹˜ æ–¹å¼ä¸‰
var6 = jt.multiply(var1, var2)
print("var6: \n", var6)
```

* **ä¸¤ä¸ª Var çŸ©é˜µç›¸ä¹˜ï¼š**
